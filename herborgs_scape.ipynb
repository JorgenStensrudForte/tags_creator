```python
import requests
from PyPDF2 import PdfFileReader
import pandas as pd
import ast

df = pd.read_excel("./data/data_sheets.xlsx")
```

```python
df['Data sheets'] = df['Data sheets'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
substring_166 = df.loc[166, "Data sheets"][0][25:]
df.loc[166, "Data sheets"] = [substring_166]
substring_167 = df.loc[167, "Data sheets"][0][25:]
df.loc[167, "Data sheets"] = [substring_167]
```

```python

df = df.dropna(subset=['Data sheets'])
df = df[df['Data sheets'].apply(lambda x: len(x) > 0 if isinstance(x, list) else False)]
df['Data sheets'] = df['Data sheets'].apply(lambda x: [x[0]] if len(x) > 1 else x)
df
```

```python
from urllib.parse import urlparse
```

```python
# Initialize an empty list to store filenames
filenames = []

for url in df["Data sheets"]:
    print(type(url))
    
    parsed_url = urlparse(url[0])
    response = requests.get(url[0])

    # Extract the filename and append to the list
    filename = parsed_url.path.split("/")[-1]
    filenames.append(filename)

    # Ensure the request was successful
    if response.status_code == 200:
        print(url[0])
        with open('./data_sheets/' + filename, 'wb') as f:
            f.write(response.content)
    else:
        print("Failed to retrieve the document")

# Add the filenames as a new column to the DataFrame
df['Filename'] = filenames
```

```python
import PyPDF2
import os
for file in os.listdir("data_sheets"):
 
    # creating a pdf file object
    pdfFileObj = open('./data_sheets/' + file, 'rb')
    print(file)
    
    # creating a pdf reader object
    pdfReader = PyPDF2.PdfReader(pdfFileObj)
    
    # printing number of pages in pdf file
    count = 0
    for page in pdfReader.pages:
        count += 1
        if "features" in page.extract_text().lower():

            index = page.extract_text().lower().index("features")
            page_text = page.extract_text()[index:]
            # button_index =  page_text.index("•")
            # Splitting the text into lines
            lines = page_text.split('\n')

            # Extracting bullet points and handling multi-line bullet points
            bullet_points = []
            current_point = ""
            for line in lines:
                if line.strip().startswith('•'):
                    if current_point:
                        bullet_points.append(current_point.strip())
                    current_point = line.strip('• ')
                elif current_point:
                    current_point += " " + line.strip()

            # Adding the last point if it exists
            if current_point:
                bullet_points.append(current_point.strip())

            # Printing the extracted list
            # for index, bullet in enumerate(bullet_points, start=1):
            #     print(f"{index}. {bullet}")
            print(bullet_points)

    
    # closing the pdf file object
    pdfFileObj.close()
```
