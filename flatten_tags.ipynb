```python
import pandas as pd
import json

# Load the products data
products = pd.read_excel("data/12_15/All Products (5).xlsx")
print(products.columns)
```

```python
for index, row in products.iterrows():
    if not pd.isnull(row["product_tags"]):
        # Split the string into parts by comma
        parts = row["product_tags"].split(',')
        # Filter out 'nan' parts
        parts = [part for part in parts if part.strip() != 'nan']
        # Join the parts back together with comma
        product_tags_string = '[' + ','.join(parts) + ']'
        try:
            product_tags = json.loads(product_tags_string)
            product_tags = [list(tag.values())[0] for tag in product_tags]
            products.at[index, "product_tags"] = product_tags
        except json.decoder.JSONDecodeError:
            print("Error decoding product tags for product :", index)
            print(product_tags_string)
    
```

```python
products.to_excel("data/12_15/all_products_6.xlsx", index=False)
```

```python
for index, row in products.iterrows():
    if not pd.isnull(row["Block2"]):
        row_test = row["Block2"]
        try:
            row_test = json.loads(row_test)
        except json.decoder.JSONDecodeError as e:
            print(f"Error decoding product tags for product at index {index}: {e}")
            error_position = e.pos
            print(f"String around the error position: {row_test[max(0, error_position-10):error_position+10]}")
```
