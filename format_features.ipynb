```python
import pandas as pd

# Load the products data
products = pd.read_excel("data/12_15/All Products (4).xlsx")
```

```python
def modify_features(features):
    output = features
    # If there are no line breaks, replace periods with semicolons
    if pd.isnull(features):
        return ""
    # output = re.sub(r'- ', '', features)

    
    output = output.replace('\n', ';')
    # output = re.sub(r'(?<!:)\s-\s(?! \d)', ';', output)




    # Replace new line characters with semicolons
    if ";" not in output:
        output = output.replace('.', ';')

    output = output.split(';')
    output = [x.strip() for x in output]
    output = [x + ';' for x in output]
    output = ''.join(output)

    output = re.sub(r'(\d)\s-\s(\d)', r'\1-\2', output)
    output = re.sub(r'(\w)\s-\s(\d)', r'\1-\2', output)
    output = output.replace('- ', ';')
    # output = re.sub(r'- ', ';', output)
    # output = re.sub(r'(?<!:)(?<!\d)\s-\s(?! \d)', ';', output)
    # output = re.sub(r'(?<!\s)-\s', ';', output)
    # # output = re.sub(r'(^|;\s)-\s', ';', output)
    # output = re.sub(r'(?<=;)\s-\s', ';', output)

    # Split the string into a list of items
    output = output.split(';')
    # Remove leading and trailing whitespace from each item
    output = [x.strip() for x in output]
    # Remove empty items from the list
    output = [x for x in output if x != '']
    # Remove • and - from the strings in the list
    output = [x.replace('•', '') for x in output]
    # output = [x.replace('- ', '') for x in output]
    output = [x.strip() for x in output]
    # Add a semicolon to the end of each item
    output = [x + ';' for x in output]
    # Join the list into a single string with a newline after each item
    output = '\n'.join(output)

    return output

# # Modify 1 element
# feature_1 = products['Features'][2]
# print(feature_1)
# output = (modify_features(feature_1))
# print("_________")
# print(output)
for x in range(250):
    print("_________")

    feature_1 = products['Features'][x]
    # print("Feature:", x," - ", feature_1)
    output = (modify_features(feature_1))
    print("_________")
    print("output: ", x," - ",output)
```
