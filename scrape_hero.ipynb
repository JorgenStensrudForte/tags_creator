```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df = pd.read_excel("./data/all_products_updated.xlsx")
df
```

```python
def find_image(url):
    response = requests.get(url)
    html = response.text

    # Parse the HTML
    soup = BeautifulSoup(html, 'html.parser')

    hero_container = soup.find("div", "Hero__container")

    # # Check for <img> tag within the 'Hero__container'
    # img_tag = hero_container.find('img') if hero_container else None

    hero_img = []
    other_img = []

    if soup.find("img"):
        for img in soup.find_all("img"):
            print(img)
            if hero_container and hero_container.find("img", src=img["src"]):
                hero_img.append("https://www.kongsberg.com" + img['src'])
            else:
                other_img.append("https://www.kongsberg.com" + img['src'])
            
    print(url)
    print("Hero", hero_img)
    print("Other", other_img)

    hero_img_value = None if not hero_img else hero_img
    other_img_value = None if not other_img else other_img

    img_list = [hero_img_value, other_img_value]

    return img_list
```

```python
df[["Hero Image", "Other Image"]] = df.apply(lambda row: pd.Series(find_image(row["url"])) if pd.notna(row["url"]) or row["url"] == [[],[]] else pd.Series([None, None]), axis=1)
```

```python
df.to_excel("./data/AI_products_hero.xlsx")
```
