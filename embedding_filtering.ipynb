```python
df_filters = pd.read_excel("filter_tags_12_12.xlsx")

df_filters
```

```python
hierarchy = {}
for category, group in df_filters.groupby('category'):
    hierarchy[category] = {}
    for _, row in group.iterrows():
        prod_type =  row['technology']
    hierarchy[category] = prod_type
```

```python
hierarchy
```

```python
hierarchy["Autonomous and uncrewed solutions"]
```

```python
df = filtered_df.copy()
```

```python
df
```

```python
df["Product category"]
```

```python
text = df["General text"].iloc[4] + df["Features"].iloc[4] + df["Technical Specifications"].iloc[4]
```

```python
text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

filters = hierarchy["Autonomous and uncrewed solutions"].split(",") + hierarchy["Surface navigation & positioning"].split(",") + hierarchy["Surveillance & monitoring"].split(",")

# Combining the text and categories for TF-IDF vectorization
combined_texts = [text] + filters

# Creating the TF-IDF Vectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(combined_texts)

# Calculating cosine similarity between the text and each category
cosine_similarities = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()

# Mapping each category with its similarity score
category_similarity = dict(zip(filters, cosine_similarities))

# Sorting categories based on similarity score
sorted_category_similarity = sorted(category_similarity.items(), key=lambda item: item[1], reverse=True)

sorted_category_similarity

```
